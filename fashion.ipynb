```python
# Sütun isimlerini temizle
fashion.columns = fashion.columns.str.strip()

columns_to_drop = ['Section 2: Style Preferences', 'Column 18', 'Timestamp', '3.Profession', '17. From scale 1-10 how much do you think your clothing style reflects about your personality?', '14. How active is your daily lifestyle?', 'Section 5: Personal Preferences\n15. How important is comfort in your clothing choices']
data = fashion.drop(columns=columns_to_drop)

# Temizlenmiş veri setinin ilk birkaç satırını göster
print(data.head())
```

```python
# Fonksiyon: Eksik değerleri analiz et
#bütün sütunlarda eksik gözlem sayısını ve yüzdesini hesaplıyor.
#Sadece eksik değeri > 0 olan sütunları listeliyor.
def missing_values(data, name="Veri Seti"):
    missing = data.isnull().sum()
    percent = (missing / len(data)) * 100
    result = pd.DataFrame({'Eksik Değer Sayısı': missing, 'Yüzde (%)': percent})
    result = result[result['Eksik Değer Sayısı'] > 0]
    print(f"\n{name} Eksik Değerler:\n", result)

missing_values(data, "data")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Kategorik sütun isimleri
columns = [
    "1.Age Group",
    "2.Gender",
    "5. What’s your favorite color palette for clothing?",
    "6. Do you prioritize functionality or aesthetics in your outfits?",
    "7.Which of these best describes your wardrobe?",
    "9.What influences your clothing purchases the most?",
    "10. Where do you typically shop for clothes? (Select all that apply)",
    "12.Do you often experiment with new styles or stick to what you know?",
    "13. What kind of footwear do you wear most often?",
    "16.If you had to choose, would you prefer timeless pieces or trendy items?"
]

def plot_top_categories(df, column_name, top_n=5, palette_name="tab10"):
    counts = df[column_name].value_counts().head(top_n)
    colors = sns.color_palette(palette_name, n_colors=len(counts))
    
    plt.figure(figsize=(10,5))
    bars = plt.bar(counts.index, counts.values, color=colors)
    plt.title(f"{column_name} - Top {top_n} Categories")
    plt.xlabel(column_name)
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
for col in columns:
    # Eksik veri varsa uyarı verelim
    if col not in data.columns:
        print(f"Uyarı: '{col}' sütunu veri setinde yok!")
        continue
    plot_top_categories(data, col)
```

```python
# Grafik stili
sns.set(style="whitegrid")

# Sütun isimleri kısa ve kolay olsun:
data_subset = data[["Section 3: Shopping Habits\n8. How often do you shop for new clothes?", "2.Gender", "1.Age Group"]].dropna()

# Yaş gruplarını sırala
age_order = sorted(data_subset["1.Age Group"].unique())
# Alışveriş sıklığı için mantıklı sıra belirleyelim
shop_order = data_subset["Section 3: Shopping Habits\n8. How often do you shop for new clothes?"].value_counts().index.tolist()

# Cinsiyetler (renk için)
gender_order = data_subset["2.Gender"].unique()

# FacetGrid ile yaş grubuna göre ayırıyoruz
g = sns.catplot(
    data=data_subset,
    x="Section 3: Shopping Habits\n8. How often do you shop for new clothes?",
    hue="2.Gender",
    col="1.Age Group",
    kind="count",
    col_wrap=4,
    order=shop_order,
    hue_order=gender_order,
    palette="Set2",
    height=4,
    aspect=1
)

g.set_titles("{col_name}")
g.set_axis_labels("Shopping Frequency", "Count")
for ax in g.axes.flatten():
    ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
